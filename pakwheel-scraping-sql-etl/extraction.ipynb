```python
# Saare cards
cards = soup.find_all("div", class_="col-md-9 grid-style")  # exact div class adjust if needed

final_data = []

for card in cards:

    # ---------- Title ----------
    try:
        title = card.find("h3").text.strip()
    except:
        title = None

    # ---------- Ratings ----------
    try:
        ratings = card.find("span", class_="rating-stars").text.strip()
    except:
        ratings = None

    # ---------- Price ----------
    try:
        price = card.find("div", class_="price-details").text.strip()
    except:
        price = None

    # ---------- UL Details ----------
    try:
        ul = card.find("ul", class_="list-unstyled search-vehicle-info-2 fs13")
        li_values = [li.text.strip() for li in ul.find_all("li")]
    except:
        li_values = []

    # Extract UL-based values safely
    year          = li_values[0] if len(li_values) > 0 else None
    mileage       = li_values[1] if len(li_values) > 1 else None
    fuel          = li_values[2] if len(li_values) > 2 else None
    engine        = li_values[3] if len(li_values) > 3 else None
    transmission  = li_values[4] if len(li_values) > 4 else None
    grade         = li_values[5] if len(li_values) > 5 else None

    # ---------- Final dictionary ----------
    car_data = {
        "title": title,
        "ratings": ratings,
        "price": price,
        "year": year,
        "mileage": mileage,
        "fuel": fuel,
        "engine": engine,
        "transmission": transmission,
        "grade": grade
    }

    final_data.append(car_data)

# Print final structured list
for car in final_data:
    print(car)
```

```python
df = pd.DataFrame(final_data)
```

```python
import time

# Basic setup
HEADERS = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
    'Accept-Language': 'en-US, en;q=0.5'
}

base_url = "https://www.pakwheels.com/used-cars/karachi/24857?page={}"

all_cars = []   # Store all scraped data

for page in range(1, 11):  # 1 to 10 pages
    print(f"Scraping page {page}...")
    url = base_url.format(page)
    
    r = requests.get(url, headers=HEADERS)
    soup = BeautifulSoup(r.content, "html.parser")
    
    cards = soup.find_all("div", class_="col-md-9 grid-style")
    
    for card in cards:
        
        # Extract title
        title_tag = card.find("h3")
        title = title_tag.text.strip() if title_tag else None
        
        # Extract ratings
        rating_tag = card.find("span", class_="rating-stars")
        ratings = rating_tag.text.strip() if rating_tag else None
        
        # Extract price
        price_tag = card.find("div", class_="price-details")
        price = price_tag.text.strip() if price_tag else None
        
        # Extract UL Details
        ul = card.find("ul", class_="list-unstyled search-vehicle-info-2 fs13")
        
        if ul:
            li = [x.text.strip() for x in ul.find_all("li")]
        else:
            li = []
        
        car_data = {
            "title": title,
            "ratings": ratings,
            "price": price,
            "year": li[0] if len(li) > 0 else None,
            "mileage": li[1] if len(li) > 1 else None,
            "fuel": li[2] if len(li) > 2 else None,
            "engine": li[3] if len(li) > 3 else None,
            "transmission": li[4] if len(li) > 4 else None,
            "grade": li[5] if len(li) > 5 else None
        }
        
        all_cars.append(car_data)
    
    time.sleep(1)  # Prevent getting blocked

print("Scraping complete!")
```

```python
df = pd.DataFrame(all_cars)

# Save as JSON
df.to_json("pakwheels_10_pages.json", orient="records", lines=True)

print("JSON Saved: pakwheels_10_pages.json")
```
